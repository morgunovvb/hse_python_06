* Программа перевода содержимого файла в словарь

```python
import time
import json

with open('Data/purchase_log.txt', encoding='utf-8') as file:
    lines = file.readlines()
purchases = {
    json.loads(line.strip())['user_id']: json.loads(line.strip())['category']
    for line in lines
}

count = 0
for key, value in purchases.items():
    if count >= 2:
        break
    print(f"{key} '{value}'")
    count += 1
```

* Программа склеивания файлов

```python
import os
import csv
import json
import time

# Чтение purchase_log.txt и формирование словаря покупок
purchases = {}
with open('Data/purchase_log.txt', encoding='utf-8') as log_file:
    for line in log_file:
        record = json.loads(line.strip())
        normalized_user_id = record['user_id'].strip('"\' ').lower()  # Приводим к стандартному виду
        purchases[normalized_user_id] = record['category']

count = 0
print(f'Всего элементов в словаре {len(purchases)}')
for key, value in purchases.items():
    if count >= 2:
        break
    print(f"{key} '{value}'")
    count += 1

# Проверка и удаление старого файла funnel.csv
if os.path.exists('Data/funnel.csv'):
    os.remove('Data/funnel.csv')
    print(f'Файл удален, результат проверки {os.path.exists('Data/funnel.csv')}')

# Обработка visit_log.csv
with open('Data/visit_log.csv', mode='r', newline='', encoding='utf-8-sig') as visit_file,\
     open('Data/funnel.csv', mode='w', newline='', encoding='utf-8') as funnel_file:
    
    reader = csv.DictReader(visit_file)
    writer = csv.writer(funnel_file)
    
    # Запись заголовков
    writer.writerow(['user_id', 'source', 'category'])
    
    # Обход каждой строки visit_log.csv
    for row in reader:
        user_id = row.get('user_id', '').strip('"\' ').lower()
        
        # Проверяем наличие user_id в словаре покупок
        if user_id in purchases:
            category = purchases[user_id]
            writer.writerow([user_id, row['source'], category])
            print(f"Дописано: {user_id}, {row['source']} => {category}")  # Сообщение для отладки

# Вывод первых трёх строк файла funnel.csv
with open('Data/funnel.csv', mode='r', newline='', encoding='utf-8') as funnel_file:
    reader = csv.reader(funnel_file)
    headers = next(reader)  # Пропускаем заголовок
    count = 0
    for row in reader:
        if count >= 3:
            break
        print(', '.join(row))
        count += 1
```
